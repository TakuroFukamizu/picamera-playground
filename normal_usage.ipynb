```python
import sys
from picamera import PiCamera

# at once
cam = PiCamera()
```

```python
from IPython.display import Image, display_jpeg

# config camera
width = 800
height = 600
cam.resolution = (width, height)

def take_photo_and_preview(filename):
    cam.capture(filename) 
    display_jpeg(Image(filename))
```

```python
# normal range

print("NORMAL MODE")
take_photo_and_preview("photo_normal.jpg")
```
